# Unmixing the image

The model we are using to decompose the image into the sum of the contributions of the endmembers is:

\begin{equation*}
\mathbf{X} = \mathbf{S}\mathbf{A} + \mathbf{E} 
\end{equation*}

 $\mathbf{X}\in \mathbb{R}^{L\times N}$ : data matrix
 
 $\mathbf{S}\in \mathbb{R}^{L\times P}$ : endmember matrix
 
 $\mathbf{A}\in \mathbb{R}^{P\times N}$ : abundance matrix
 
 $\mathbf{E}\in \mathbb{R}^{L\times N}$ : additive noise matrix
 
 where $L$ is the number of spectral bands, $P$ is the number of materials, and $N$ is the number of pixels
 
 The hyperspectral image is represented as a matrix with $N$ pixels, so the first thing to do is to reshape the image to a big matrix.

We can try and estimate the abundance matrix using a least squares formulation:

\begin{equation*}
\hat{A} = \underset{\mathbf{A}>\mathbf{0}, \mathbf{1}_{P}^T \mathbf{A} = \mathbf{1}_{N}^{T} }{\textrm{arg min}}\frac{1}{2}||\mathbf{X}-\mathbf{SA}||_{F}^{2} = \underset{\mathbf{A}>\mathbf{0}, \mathbf{1}_{P}^T \mathbf{A} = \mathbf{1}_{N}^{T} }{\textrm{arg min}}\frac{1}{2} \sum_{i=1}^{N} ||\mathbf{x}_{n}-\mathbf{S}\mathbf{a}_{n}||_{2}^{2}
\end{equation*}

where $\mathbf{1}_{.}$ is a vector of ones whose size is given in index 

Choose an algorithm to solve the optimization problem, and implement it both in a serial version and a version parallelized across the pixels.

```julia
# Acquire processes for computation, then cast everything to the processes
CPU_CORES = 8;

addprocs(CPU_CORES - 1);
```

```julia
#cast necessary functions and imports everywhere
@everywhere using LinearAlgebra
@everywhere f(S,A,X) = 1/2 * norm(X - S*A,2);
@everywhere ∇f(S,A,X)  = 2 * transpose(S)*S*A .- transpose(S)*X; # I only take a derivative with respect to A because I believe it's the only variable
```

# Parameters
## Parameter benchmark
We'll be testing parameters to find the best parameters to use for speed purposes.  
The 2 parameters we'll be looking at are : 
- number of processes used for the algorithm. In a perfect world, the speed up of the algorithm should be a linear function of the number of processes used.
- batch size : This is the size of the data patches sent to the individual processes. This has an impact because in reality, the time to transfer data to CPU cores in not neglectable, so minimizing it is best.

```julia
height = 300;
width = 300;
precision = 10e-8;
max_iter = 100;
step_size = 2/opnorm(transpose(sources)*sources);

#########################################

N = height*width
A = reshape([rand(4,1) for i=1:N],1,N);

img_small = img[1:height,1:width,:]; 
```

```julia
img_small = permutedims(img_small, (3,1,2));
img_small_flatt = reshape(img_small, 255, N);
img_small_para = reshape([img_small_flatt[:,i] for i=1:N],1,N);
```

```julia
### Serial version of the algorithm
@everywhere function GD_Pixel(X,A,S,precision,max_iter,step_size)
    
    nb_it = 0
    while norm(∇f(S,A,X)) > precision && nb_it < max_iter
        
        A = A - step_size*∇f(S,A,X);
        A = max.(A,0); #We don't allow negative value
        nb_it+=1;
    end
    for i in 1:size(A)[2]
        if sum(A[i]) != 1 #Because of clipping
            A[i] = A[i]/sum(A[i]);
        end
    end
    return A;
    
end


#Transform every parameters in array of dime (1,N) to use map and pmap functions
sources_para = reshape([sources for p=1:N],1,N);
precision_para = precision * ones(1,N);
max_iter_para = max_iter * ones(1,N);
step_size_para = step_size * ones(1,N);
```

```julia
#Compute A using only one proc
A = @time map(GD_Pixel,img_small_para,A,sources_para,precision_para,max_iter_para,step_size_para);  
```

```julia
@everywhere function get_img(A,S)
    A_pred = zeros(4,N);
    for i=1:size(A,2)
        A_pred[:,i] = A[i];
    end
    X_pred = S*A_pred;
    X_pred = reshape(X_pred,size(X_pred,1),height,width);
    X_pred = permutedims(X_pred,(2,3,1));
    return A_pred,X_pred
end

A_pred,img = get_img(A,sources);
#visualize a spectral band on its own
sb = 10
p2 = heatmap(img[:,:,sb],aspect_ratio=1, yflip = true,c=:ice)
```

```julia
A = reshape([rand(4,1) for i=1:N],1,N);
A = @time pmap(GD_Pixel,img_small_para,A,sources_para,precision_para,max_iter_para,step_size_para);  
```

```julia
#visualize a spectral band on its own
A_pred,img = get_img(A,sources);
sb = 10
p2 = heatmap(img[:,:,sb],aspect_ratio=1, yflip = true,c=:ice)
```

```julia
function generate_rgb_img(img)
    rgb = zeros(size(img,1),size(img,2),3)  
    rgb[:,:,3] = img[:,:,67] * 1.3520674485645392; 
    rgb[:,:,2] = img[:,:,109] * 1.194526746571833;
    rgb[:,:,1] = img[:,:,141] * 1.092957082800559; 
    return rgb
end
```

```julia
# Define the benchmarking suite

suite = BenchmarkGroup()

for n_workers in 1:nprocs()
    worker_pool = WorkerPool(procs()[1:n_workers])
    suite[n_workers] = BenchmarkGroup(["$n_workers processes"])
    for batch_size in 1:N/(20*n_workers):(N/n_workers)
        b_size = round(batch_size)
        suite[n_workers][b_size] = @benchmarkable pmap(GD_Pixel,$worker_pool,img_small_para,A,sources_para,precision_para,max_iter_para,step_size_para, batch_size=$b_size) seconds=120 samples=10
    end
end
println(suite)
```

```julia
tune!(suite, verbose=true)
```

```julia
# Run the benchmark
println(suite)
results = run(suite, verbose = true)
```

```julia
using Serialization

# Serialize the results to not have to run it again
open(f -> serialize(f,results), "benchmark_results.jls", "w")

# Serialize the benchmark group 
open(f -> serialize(f,suite), "benchmark_suite.jls", "w")
```

## Result analysis

Let's take a look at the results of the benchmark

```julia
# open the result file
file_name = "benchmark_results.jls"
benchmark_results = open(deserialize, file_name)
```

```julia
using StatsPlots

# Extract the data from the benchmarkGroup, sorting the trials by the key value 
function extract_data(benchmark_group::BenchmarkGroup) 
    # Get the (keys,Trial) pairs of the benchmark group
    indexes = []
    trials = []
    for trial in benchmark_group
        push!(indexes,trial[1])
        push!(trials,trial[2])
    end
    perm = sortperm(indexes)
    indexes=indexes[perm]
    trials=trials[perm]
    
    return indexes,trials
end
        
```

## Optimal batch size by n_procs
Intuition tells us that as long as we have enough memory, we should take batch_size = n_samples/n_procs.  
Let's see if this is the case.

```julia
plots = []
indexes, trial_groups = extract_data(benchmark_results)
for (n_procs,trial_group) in zip(indexes,trial_groups)
    indexes,trials = extract_data(trial_group)
    # Find the best time for each trial
    line = plot(indexes,map(v->minimum(v.times),trials), label="recorded_times", title="$n_procs processes")
    v_25 = plot!(line,[N/(4*n_procs)],label="25% of N/n_procs", st=:vline)
    v_50 = plot!(line,[N/(2*n_procs)],label="50% of N/n_procs", st=:vline)
    v_75 = plot!(line,[(3*N)/(4*n_procs)],label="75% of N/n_procs", st=:vline)

    push!(plots,line)
end

n_plots = length(plots)
plot(plots...,layout=(n_plots,1),size=(600,600*n_plots))
```

## Speedup analysis
Let's see how the speedup behaves as a function of the number of procs.   
In a perfect world, speedup(n_procs) = n_procs.

```julia
optimal_batch_sizes = []
minimum_times = []
indexes, trial_groups = extract_data(benchmark_results)
for (n_procs,trial_group) in zip(indexes,trial_groups)
    batch_sizes,trials = extract_data(trial_group)
    trial_minimums = map(v->minimum(v.times), trials)
    (minimum_time,index) = findmin(trial_minimums)
    
    push!(optimal_batch_sizes, batch_sizes[index])
    push!(minimum_times, minimum_time)
end

speedup = minimum_times[1]*ones(size(minimum_times)) ./minimum_times
p = plot(indexes, speedup, label=optimal_batch_sizes, st=:scatter)
x = 1:1:8
plot!(p,x,x,label="y=x")
```

### Optimal batch_size as a percentage of N/n_procs

```julia

best_batch_size = optimal_batch_sizes ./ ((N * ones(length(optimal_batch_sizes))) ./ indexes)
plot(indexes,best_batch_size, st=:scatter)
```
